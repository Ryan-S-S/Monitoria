```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('Datasets/movies_metadata.csv', low_memory = False)
movies.head(3)
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head(3)
```

```python
# Pre-processamento dos filmes - Filtrando apenas as colunas necessÃ¡rias
movies = movies[['id', 'original_title', 'original_language', 'vote_count']]
movies.rename(columns = {'id' : 'ID_MOVIE', 'original_title': 'TITLE', 'original_language': 'LANGUAGE', 'vote_count' : 'VOTE_COUNT'}, inplace=True)
movies.head(3)
```

```python
# Verificando valores nulos - Filmes
print('Quantidade de Valores Nulos por Coluna : \n')
print(movies.isna().sum())
print('Removendo linhas com valores nulos')
movies.dropna(inplace=True)
print('Quantidade de Valores Nulos por Coluna apÃ³s a operaÃ§Ã£o: \n')
print(movies.isna().sum())
```

```python
# Filmes com mais de 999 avaliaÃ§Ãµes
movies = movies[movies['VOTE_COUNT'] > 999]
```

```python
movies_language = movies['LANGUAGE'].value_counts()
movies_language.head(10)
```

```python
movies = movies[movies['LANGUAGE'] == 'en']
```

```python
movies.info()
```

```python
# ConversÃ£o dos Object para inteiro para cruzar dados com ratings
movies['ID_MOVIE'] = movies['ID_MOVIE'].astype(int)
```

```python
movies.shape
```

```python
# Pre-processamento das avaliaÃ§Ãµes - Filtrando apenas as colunas necessÃ¡rias
ratings = ratings[['userId', 'movieId', 'rating']]
ratings.rename(columns ={'userId' : 'ID_USER', 'movieId' : 'ID_MOVIE', 'rating' : 'RATING'} , inplace=True)
ratings.head(3)
```

```python
# Verificando valores nulos - AvaliaÃ§Ãµes
print('Quantidade de Valores Nulos por Coluna : \n')
print(ratings.isna().sum())
print('Removendo linhas com valores nulos')
ratings.dropna(inplace=True)
print('Quantidade de Valores Nulos por Coluna apÃ³s a operaÃ§Ã£o: \n')
print(ratings.isna().sum())
```

```python
ratings['ID_USER'].value_counts() # Quantidade de avaliaÃ§Ãµes por usuÃ¡rio
```

```python
# Selecionado apenas usuÃ¡rio que fizeram mais de 999 avaliaÃ§Ãµes
qt_ratings = ratings['ID_USER'].value_counts() > 999
y = qt_ratings[qt_ratings].index
y.shape
```

```python
# Usuarios selecionados
y
```

```python
# Selecionando o user e seus dados
print("Formato das AvaliaÃ§Ãµes : ", ratings.shape)
print("Selecionando apenas os usuÃ¡rios com mais de 999 avaliaÃ§Ãµes")
ratings = ratings[ratings['ID_USER'].isin(y)]
print("Formato das AvaliaÃ§Ãµes : ", ratings.shape)
```

```python
ratings.info()
```

```python
# Juntandos os dataframes
ratings_and_movies = ratings.merge(movies, on = 'ID_MOVIE')
ratings_and_movies.head(3)
```

```python
print("Formato do dataframe mesclado : ", ratings_and_movies.shape)
print("Valores Nulos no df mesclado : ", ratings_and_movies.isna().sum())
```

```python
# Removendo valores duplicados de avaliaÃ§Ãµes
ratings_and_movies.drop_duplicates(['ID_USER', 'ID_MOVIE'], inplace=True)
ratings_and_movies.shape
```

```python
# Excluindo a variavel ID_MOVIE
del ratings_and_movies['ID_MOVIE']
ratings_and_movies.head(3)
```

```python
movies_pivot = ratings_and_movies.pivot_table(columns = 'ID_USER', index = 'TITLE', values = 'RATING')
movies_pivot.head(3)
```

```python
movies_pivot.fillna(0, inplace=True)
movies_pivot.head(3)
```

```python
from scipy.sparse import csr_matrix
movies_sparse = csr_matrix(movies_pivot)
```

```python
type(movies_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
model.fit(movies_sparse)
```

```python
# Previsao
distances, sugestions = model.kneighbors(movies_pivot.filter(items=['Toy Story'], axis=0).values.reshape(1,-1))

for i in range(len(sugestions)) :
  print(movies_pivot.index[sugestions[i]])
```
